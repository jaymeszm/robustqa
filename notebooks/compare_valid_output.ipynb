```python
import pandas as pd
```

```python
df = pd.read_csv('validation_output.csv')
df.head()
```

```python
len(df)
```

### 0. answer == baseline, answer != mlmqa_alt_data_aug

What does the combined system fail to get compared to the baseline?

```python
df0 = df.loc[(df.answer == df.baseline) & (df.answer != df.mlmqa_alt_data_aug)] 
df0[['id', 'question', 'context', 'answer', 'baseline', 'mlmqa_alt_data_aug']]
```

```python
len(df0) / len(df)
```

### 1. answer == mlmqa_alt == data_aug; answer != mlmqa_alt_data_aug

What does the combined system fail to get, compared to the underlying systems?

```python
df1 = df.loc[(df.answer == df.mlmqa_alt) & (df.answer == df.data_aug) & (df.answer != df.mlmqa_alt_data_aug)] 
df1[['id', 'question', 'context', 'answer', 'mlmqa_alt', 'data_aug', 'mlmqa_alt_data_aug']]
```

```python
len(df1)/len(df)
```

### 2. answer != mlmqa_alt, answer != data_aug; answer == mlmqa_alt_data_aug

Does the combined system retrieve anything that the individual systems fail to get?

```python
df2 = df.loc[(df.answer != df.mlmqa_alt) & (df.answer != df.data_aug) & (df.answer == df.mlmqa_alt_data_aug)] 
df2[['id', 'question', 'context', 'answer', 'mlmqa_alt', 'data_aug', 'mlmqa_alt_data_aug']]
```

```python
len(df2)/len(df)
```
